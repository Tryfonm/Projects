```python
import numpy as np
from matplotlib import pyplot as plt
from myOptim import myUnconstrainedOptimizer
```

```python
# -------------------------- Example 1 -------------------------- #

def obj_toy(xy):
    x = xy[0].item()
    y = xy[1].item()
    return np.array([[x**2 + 2*y**2]])  

# The optimizer expexts a function whose output shape is of  [Nf, 1] 
x0 = np.array([[22],[18]])
opt = myUnconstrainedOptimizer()
```

```python
x = opt.minimize_grad_desc(obj_toy, x0, searchMethod='lineSearch', search_parameters=[0.01, 0.6])
opt.plot2d(lims=[-30,30,-25,25])
```

```python
x = opt.minimize_bfgs(obj_toy, x0, searchMethod='lineSearch', search_parameters=[0.01, 0.6])
opt.plot2d(lims=[-30,30,-25,25])
```

```python
# -------------------------- Example 2 -------------------------- #

def obj_rosenbrock(xy):
    x = xy[0].item()
    y = xy[1].item()
    return np.array([[(1 - x)**2 + 100*(y - x**2)**2]])

# The optimizer expexts a function whose output shape is of  [Nf, 1] 
x0 = np.array([[-1.5],[1]])
opt = myUnconstrainedOptimizer()
```

```python
x = opt.minimize_grad_desc(obj_rosenbrock, x0, searchMethod='lineSearch', search_parameters=[0.01, 0.6])
opt.plot2d(contourDensity=200, lims=[-3,3,-1.5,3], figsize=(24,12))
```

```python
x = opt.minimize_bfgs(obj_rosenbrock, x0, searchMethod='lineSearch', search_parameters=[0.01, 0.6])
opt.plot2d(lims=[-3,3,-1.5,3], figsize=(24,12))
```
